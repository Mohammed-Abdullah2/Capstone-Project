```python
import pandas as pd

df2 = pd.DataFrame()  # DataFrame نهائي لتخزين النتائج
years = list(map(str, range(2016, 2025)))  # السنوات المطلوبة (2016-2024)

for i in range(0, index):  # تكرار لعدد معين من الصفوف
    temp = df.loc[i, :]
    overviewVideoViews_date = temp['overviewVideoViews_date']
    overviewVideoViews_count = temp['overviewVideoViews_count']

    # إنشاء DataFrame للبيانات
    dfa = pd.DataFrame({
        'overviewVideoViews_date': overviewVideoViews_date,
        'overviewVideoViews_count': overviewVideoViews_count
    })

    # حساب الفرق السنوي بنفس المنطق الموجود في الكود الأصلي
    di = {}
    for k in range(dfa.shape[0]):
        date = dfa.loc[k, 'overviewVideoViews_date']
        count = int(dfa.loc[k, 'overviewVideoViews_count'])
        year = date[:4]
        di[year] = count

    di_key = list(di.keys())
    di_value = list(di.values())
    di_a = {}
    for j in range(len(di_key) - 1, -1, -1):
        if j > 0:
            di_a[di_key[j]] = abs(di_value[j] - di_value[j - 1])
        else:
            di_a[di_key[j]] = di_value[j]

    # ضمان وجود جميع السنوات في النتيجة النهائية
    output = {f"{year}_views": di_a.get(year, 0) for year in years}

    # تحويل النتائج إلى شكل DataFrame
    df_1 = pd.DataFrame({
        'profile_name': [temp['profile_name']],  # إضافة اسم الملف
        **output
    })

    # دمج البيانات الجديدة مع DataFrame النهائي
    df2 = pd.concat([df2, df_1], axis=0, ignore_index=True)

# النتيجة النهائية
df2.head()
```

```python
import pandas as pd

df3 = pd.DataFrame()  # DataFrame نهائي لتخزين النتائج
years = list(map(str, range(2016, 2025)))  # السنوات المطلوبة (2016-2024)

for i in range(0, index):  # تكرار لعدد معين من الصفوف
    temp = df.loc[i, :]
    overviewVideoViews_date = temp['overviewSubscribers_date']
    overviewVideoViews_count = temp['overviewSubscribers_count']

    # إنشاء DataFrame للبيانات
    dfa = pd.DataFrame({
        'overviewSubscribers_date': overviewVideoViews_date,
        'overviewSubscribers_count': overviewVideoViews_count
    })

    # حساب الفرق السنوي بنفس المنطق الموجود في الكود الأصلي
    di = {}
    for k in range(dfa.shape[0]):
        date = dfa.loc[k, 'overviewSubscribers_date']
        count_str = dfa.loc[k, 'overviewSubscribers_count']

        # التحقق من القيم الفارغة واستبدالها بـ 0
        count = int(count_str) if count_str.strip() else 0

        year = date[:4]
        di[year] = count

    di_key = list(di.keys())
    di_value = list(di.values())
    di_a = {}
    for j in range(len(di_key) - 1, -1, -1):
        if j > 0:
            di_a[di_key[j]] = abs(di_value[j] - di_value[j - 1])
        else:
            di_a[di_key[j]] = di_value[j]

    # ضمان وجود جميع السنوات في النتيجة النهائية
    output = {f"{year}_subscribers": di_a.get(year, 0) for year in years}

    # تحويل النتائج إلى شكل DataFrame
    df_1 = pd.DataFrame({
        'profile_name': [temp['profile_name']],  # إضافة اسم الملف
        **output
    })

    # دمج البيانات الجديدة مع DataFrame النهائي
    df3 = pd.concat([df3, df_1], axis=0, ignore_index=True)

# النتيجة النهائية
df3.head()
```
